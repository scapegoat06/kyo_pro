```python
#メモ化、フィボナッチ
def fib(n):
    if n==0 or n==1:
        return 1
    else:
        return fib(n-1)+fib(n-2)


def fib_memo(n,memo={}):
    if n==0 or n==1:
        return 1
    try:
        return memo[n]
    except KeyError:
        result = fib_memo(n-1,memo)+fib_memo(n-2,memo)
        memo[n]=result
        return result

def fib_simple(n):
    fibo=[1,1]
    for i in range(2,n+1):
        fibo.append(fibo[i-1]+fibo[i-2])
    return fibo[-1]
```

```python
%time fib(20)
```

```python
%time fib_memo(2000)
```

```python
%time fib_simple(10000)
```

```python
#ナップサック問題

#全探索
n=10
W=300
item=["ポッキー","うまい棒","じゃがりこ","ベビースターラーメン",
        "チロルチョコ","かっぱえびせん","サッポロポテト","都こんぶ",
      "ヨーグルト","おにぎりせんべい"]
# ポッキー　168円　496kcal
# うまい棒　10円　45kcal
# じゃがりこ　145円　325kcal
# ベビースターラーメン 60円　347kcal
# チロルチョコ　10円　61kcal
# かっぱえびせん　124円　486kcal
# サッポロポテト 124円　446kcak
# 都こんぶ　105円 22kcal
# ヨーグレッとは入れもん　126円　110kcal
# おにぎりせんべい　184円　475kcal
w=[168,10,145,60,10,124,124,105,126,184]
v=[496,45,325,347,61,486,446,22,110,475]

def rec(i,j):
    if i==n:
        res=0
    elif j<w[i]:
        res=rec(i+1,j)
    else:
        res1=rec(i+1,j)
        res2=rec(i+1,j-w[i])+v[i]
        res=max(res1,res2)
        
            
    return res

dp=[[-1 for _ in range(W+1) ]for _ in range(n+1)]
def rec_dpmemo(i,j):
    if dp[i][j]!=-1:
        return dp[i][j]
    
    if i==n:
        res=0
    elif j<w[i]:
        res=rec_dpmemo(i+1,j)
    else:
        res1=rec_dpmemo(i+1,j)
        res2=rec_dpmemo(i+1,j-w[i])+v[i]
        res=max(res1,res2)
    
    dp[i][j]=res
            
    return res


```

```python
%time rec(0,W)
```

```python
%time rec_dpmemo(0,W)
```
